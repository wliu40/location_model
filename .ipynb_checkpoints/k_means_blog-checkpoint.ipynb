```python
%matplotlib inline
from copy import deepcopy
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
# Importing the dataset
data = pd.read_csv('xclara.csv')
print(data.shape)
data.head()
```

```python
f1 = data['V1'].values
f2 = data['V2'].values
X = np.array(list(zip(f1, f2)))
print(X.shape) ##3000行 两列
plt.scatter(f1, f2, c='black', s=7)
print(max(f1))
print(max(f2))
```

```python
# Euclidean Distance Caculator
def dist(a, b, ax=1):
    return np.linalg.norm(a - b, axis=ax)
# a = np.array([1,2])
# b = np.array([4,5])
# dist(10,1)
```

```python
np.max(X)
```

```python
# Number of clusters
k = 3
# X coordinates of random centroids
C_x = np.random.randint(0, np.max(X)-20, size=k)
# Y coordinates of random centroids
C_y = np.random.randint(0, np.max(X)-20, size=k)
C = np.array(list(zip(C_x, C_y)), dtype=np.float32)
print(C)
```

```python
# Plotting along with the Centroids
plt.scatter(f1, f2, c='#050505', s=7)
plt.scatter(C_x, C_y, marker='*', s=200, c='g')
```

```python
# To store the value of centroids when it updates
C_old = np.zeros(C.shape) ##initialize as [[0,0],[0,0],[0,0]]
# Cluster Lables(0, 1, 2)
clusters = np.zeros(len(X))
# Error func. - Distance between new centroids and old centroids
error = dist(C, C_old, None)
# Loop will run till the error becomes zero

print(C_old)
print(clusters.shape)
print(C)
print(error) #residue
```

```python
print(X[0])
print(C)
print(dist(X[0], C))
print(X[0]-C)
print(np.linalg.norm(X[0]-C, axis = 1))
print(np.argmin([1,-2,3])) ##返回-2的index： 1
```

```python
while error != 0:
    # Assigning each value to its closest cluster
    for i in range(len(X)): ##3000个instance
        distances = dist(X[i], C) #返回X[i]与C(三个中心点)的距离
        cluster = np.argmin(distances) #返回最小的那个距离的index
        clusters[i] = cluster ##assign this point to that cluster
    # Storing the old centroid values
    C_old = deepcopy(C) ##保存旧值，用于计算residue
    # Finding the new centroids by taking the average value
    for i in range(k):
        points = [X[j] for j in range(len(X)) if clusters[j] == i]
        C[i] = np.mean(points, axis=0)
    error = dist(C, C_old, None)
colors = ['r', 'g', 'b', 'y', 'c', 'm']
fig, ax = plt.subplots()
for i in range(k):
        points = np.array([X[j] for j in range(len(X)) if clusters[j] == i])
        ax.scatter(points[:, 0], points[:, 1], s=7, c=colors[i])
ax.scatter(C[:, 0], C[:, 1], marker='*', s=200, c='#050505')
```
